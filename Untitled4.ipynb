```python
import pandas as pd
import json
import matplotlib.pyplot as plt
import seaborn as sns

url = "https://raw.githubusercontent.com/ingridcristh/challenge2-data-science-LATAM/refs/heads/main/TelecomX_Data.json"
df = pd.read_json(url)

print("Valores ausentes por columna:")
missing = df.isnull().sum()
for col, val in missing.items():
    print(f"{col}: {val}")

print("Tipos de datos:")
print(df.dtypes)

print("Valores duplicados por columna:")
for col in df.columns:
    print(f"{col}: {df.duplicated([col]).sum()}")

print("Errores de formato en fechas")
for col in df.columns:
    try:
        pd.to_datetime(df[col], errors='raise')
    except:
        print(f"{col}: Sin fechas")

print("Inconsistencias detectadas en columnas")
for col in df.select_dtypes(include='object'):
    valores = df[col].astype(str).str.strip().str.lower()
    unicos = valores.unique()
    if len(unicos) > 10 or "" in unicos:
        print(f"{col} = {len(unicos)} valores 칰nicos")

cat_cols = ['Churn', 'customer', 'phone', 'internet', 'account']

for col in cat_cols:
    df[col] = df[col].astype(str).str.strip().str.lower()
    df[col] = df[col].str.replace('n o', 'no', regex=False)
    df[col] = df[col].str.replace('s i', 's칤', regex=False)
    df[col] = df[col].str.replace('si', 's칤', regex=False)
    df[col] = df[col].str.replace('nan', '', regex=False)
    df[col] = df[col].str.replace('none', '', regex=False)
    df[col] = df[col].str.replace('_', ' ', regex=False)
    df[col] = df[col].str.replace('sin servicio', 'no', regex=False)

df.replace("", pd.NA, inplace=True)
df.dropna(subset=cat_cols, inplace=True)
df.drop_duplicates(inplace=True)
df.drop_duplicates(subset='customerID', inplace=True)
df.reset_index(drop=True, inplace=True)

print("\n Valores 칰nicos tras limpieza:")
for col in cat_cols:
    print(f"{col} = {df[col].nunique()} valores 칰nicos")

print("\n Valores ausentes por columna:")
print(df.isnull().sum())

print("\n Duplicados totales:")
print(df.duplicated().sum())

def obtener_monthlycharges(x):
    try:
        x = json.loads(x.replace("'", '"'))
        return x.get('charges', {}).get('monthly')
    except:
        return None

df['MonthlyCharges'] = df['account'].apply(obtener_monthlycharges)
df['Cuentas_Diarias'] = df['MonthlyCharges'].apply(lambda x: round(x / 30, 2) if pd.notnull(x) else None)





df['Churn'] = df['Churn'].map({'s칤': 1, 'no': 0})
df.rename(columns={
    'MonthlyCharges': 'cargos_mensuales',
    'Cuentas_Diarias': 'cargos_diarios',
    'TotalCharges': 'cargos_totales'
}, inplace=True)

df['contract'] = df['account'].apply(lambda x: json.loads(x.replace("'", '"'))['contract'])
df['contract'] = df['contract'].astype('category')



num_cols = df.select_dtypes(include=['int64', 'float64']).columns

print("Estad칤sticas descriptivas por columna num칠rica:\n")
for col in num_cols:
    media = df[col].mean()
    mediana = df[col].median()
    desviacion = df[col].std()
    print(f"{col}:")
    print(f"  Media: {media:.2f}")
    print(f"  Mediana: {mediana:.2f}")
    print(f"  Desviaci칩n est치ndar: {desviacion:.2f}\n")





import ast
dict_columns = ['phone', 'internet', 'customer']
for col in dict_columns:
    df[col] = df[col].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

def traducir_phone(x):
    if x['phoneservice'] == 'no':
        return 'Sin servicio telef칩nico'
    elif x['multiplelines'] == 'no':
        return 'S칤 - Sin m칰ltiples l칤neas'
    elif x['multiplelines'] == 'yes':
        return 'S칤 - Con m칰ltiples l칤neas'
    else:
        return 'Otro'

def traducir_internet(x):
    if x['onlinesecurity'] == 'no' and x['streamingtv'] == 'no':
        return 'Sin seguridad ni streaming'
    elif x['onlinesecurity'] == 'yes' and x['streamingtv'] == 'no':
        return 'Con seguridad, sin streaming'
    elif x['onlinesecurity'] == 'no' and x['streamingtv'] == 'yes':
        return 'Sin seguridad, con streaming'
    elif x['onlinesecurity'] == 'yes' and x['streamingtv'] == 'yes':
        return 'Con seguridad y streaming'
    else:
        return 'Otro'

def traducir_customer(x):
    if x['gender'] == 'female' and x['seniorcitizen'] == 0:
        return 'Mujer - No senior'
    elif x['gender'] == 'female' and x['seniorcitizen'] == 1:
        return 'Mujer - Senior'
    elif x['gender'] == 'male' and x['seniorcitizen'] == 0:
        return 'Hombre - No senior'
    elif x['gender'] == 'male' and x['seniorcitizen'] == 1:
        return 'Hombre - Senior'
    else:
        return 'Otro'

df['phone'] = df['phone'].apply(traducir_phone)
df['internet'] = df['internet'].apply(traducir_internet)
df['customer'] = df['customer'].apply(traducir_customer)

categorical_vars = ['phone', 'internet', 'contract', 'customer']

for var in categorical_vars:
    plt.figure(figsize=(12, 6))
    sns.countplot(data=df, x=var, hue='Churn', palette='Set2')
    plt.title(f'Evasi칩n por {var}')
    plt.xticks(rotation=45, ha='right', fontsize=10)
    plt.tight_layout()
    plt.show()


df['account'] = df['account'].apply(ast.literal_eval) if isinstance(df['account'].iloc[0], str) else df['account']

df['total'] = df['account'].apply(lambda x: float(x['charges']['total']) if x['charges']['total'].strip() else None)
df['monthly'] = df['account'].apply(lambda x: float(x['charges']['monthly']) if str(x['charges']['monthly']).strip() else None)
df['tiempo_estimado'] = df.apply(lambda x: x['total'] / x['monthly'] if x['monthly'] else None, axis=1)

df['Churn'] = df['Churn'].astype('category')

plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='total', hue='Churn', kde=True, palette='Set2')
plt.title('Distribuci칩n del total gastado seg칰n Evasi칩n')
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='tiempo_estimado', hue='Churn', kde=True, palette='Set2')
plt.title('Distribuci칩n del tiempo estimado de contrato seg칰n Evasi칩n')
plt.tight_layout()
plt.show()








```

An치lisis sobre la evasi칩n de clientes (Churn)

Se realiz칩 una transformaci칩n de varias columnas que conten칤an diccionarios en formato string. Las variables 'phone', 'internet', 'customer' y 'account' fueron convertidas correctamente para poder trabajar con sus atributos internos. Tambi칠n se gestionaron valores vac칤os y se convirtieron datos num칠ricos como el total gastado y el costo mensual. Se derivaron variables adicionales como una estimaci칩n del tiempo de permanencia en funci칩n del total gastado y el pago mensual.

Se analizaron variables categ칩ricas como tipo de contrato, servicios de internet, g칠nero, edad, entre otros. Se generaron visualizaciones que muestran c칩mo se distribuye la evasi칩n entre diferentes grupos. Tambi칠n se evalu칩 la relaci칩n entre variables num칠ricas como el total gastado y el tiempo estimado de permanencia con el churn, detectando diferencias claras entre quienes abandonan y quienes permanecen.

Los clientes con contratos mensuales tienen mayor tasa de evasi칩n, al igual que aquellos sin servicios complementarios como seguridad en l칤nea o TV por streaming.
